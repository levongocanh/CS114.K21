##**import thư viện**

```
import pandas as pd
import numpy as np
import math
from sklearn.linear_model import LinearRegression
```

##**Đọc data**

```
data = pd.read_csv('/content/Book1.csv')
#del data['Unnamed: 2']
#del data['Unnamed: 3']
print(data)
data.columns = ['size', 'operations']

import matplotlib.pyplot as plt
plt.scatter(data['size'], data['operations'])
```

##**Chuẩn hóa dữ liệu**

```
X = data['size'].values.reshape(-1,1)
```

```
Y = data['operations'].values.reshape(-1,1)
```

##**Hàm tìm hệ số x**

```
def Find_X(_X):
  regr = LinearRegression()
  regr.fit(_X, data['operations'])
  return regr.coef_
```

#**Các hàm tìm hệ số x**

```
result = []
```

```
import math
from decimal import *
getcontext().prec = 9999
```

##**Hàm lg(n)**

```
X_ = np.log2(X)
```

```
for i in X:
  print(math.log2(i[0]))
```

```
re = Find_X(X_)
print(re)
result.append(re)
```

```
for i in X:
  print(re[0]*math.log2(i[0]))
```

```
sum = 0
length = len(X)
for i in range(length):
  sum+= (re[0]*math.log2(int(X[i][0]))-(Y[i][0]))**2
print(sum/length)
```

##**Hàm sqrt(n)**

```
X_ = np.sqrt(X)
```

```
for i in X:
  print((i[0]**(1/2)))
```

```
re = Find_X(X_)
print(re)
result.append(re)
```

```
for i in X:
  print(re[0]*(i[0]**(1/2)))
```
